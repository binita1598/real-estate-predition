```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,10)
```

```python
df1=pd.read_csv("MagicBricks.csv")
df1.head()
```

```python
df1.shape
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
df2=df1.drop(['area_type','Furnishing','Status'],axis='columns')
            
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3=df2.dropna()
df3.isnull().sum()
```

```python
df3.Per_Sqft.unique()
```

```python
len(df3.Locality.unique())
```

```python
df3.Locality=df3.Locality.apply(lambda x: x.strip())
Locality_stats=df3.groupby('Locality')['Locality'].agg('count')
Locality_stats
```

```python
df3.head()
```

```python
dummies=pd.get_dummies(df3.Locality)
dummies.head(3)
```

```python
df4=pd.concat([df3,dummies],axis='columns')
df4.head()
```

```python
df4.shape
```

```python
x=df4.drop('Price',axis='columns')
x.head()
```

```python
y=df4.Price
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf=LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test)
```
